```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
import matplotlib.pyplot as plt
```

`1. Create a Quantum Circuit with 2 qubits and 2 classical bits.`

`The Classical bits are used to store the measurements at the end.`

```python
qc = QuantumCircuit(2, 2)
```

`2. Add a Hadamard (H) gate to qubit 0`

`This puts qubit 0 into a superposition (it is now both 0 and 1 at the same time exist).`

```python
qc.h(0)
```

`3. Add a Controlled-X (CX or CNOT) gate.`
- Qubit 0 is the "Control" and qubit 1 is the "target".
- This means that if 1st (control) qubit is in state |1⟩, then the CX gate will flip the state of 2nd (target) qubit (from |0⟩ to |1⟩ or from |1⟩ to |0⟩). If qubit 1st (control) is in state |0⟩, then qubit 2nd (target) remains unchanged.
- This is what creates the "Entanglement" between the two qubits.

```python
qc.cx(0, 1)
```

`4. Measure the quantum bitss and store the results in the classical bits.`

```python
qc.measure([0, 1], [0, 1])
```

`5. Draw the circuit so you can see what it looks like.`

```python
print("Here is you Quantum Circuit:")
display(qc.draw("mpl")) # 'mpl' uses matplotlib for nice graphic
```

## --- RUNNING THE SIMULATION ---

`6. Get the local simulator to run the circuit.`

```python
simulator = AerSimulator()
```

`7. Run the circuit 1000 times (shots) to get statistical results.`

```python
print("\nRunning the simulation...")
job = simulator.run(qc, shots=1000)
result = job.result()
```

`8. Get the counts (how many times it resulted in 00 and 11)`

```python
counts = result.get_counts(qc)
print(f"Results after 1000 runs: {counts}")
```

`9. Visualize the results using a histogram.`

```python
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```
