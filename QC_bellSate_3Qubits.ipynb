```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
# 1. Create a Quantum Circuit with 3 qubits and 3 classical bits
qc = QuantumCircuit(3, 3)
```

```python
#  2. Apply Hadamard gate to Qubit 0 (creates the 50/50 superposition)
qc.h(0)
```

```python
# 3. Entangle Qubit 0 with Qubit 1
qc.cx(0, 1)
```

```python
# 4. NEW: Entagle Qubit 1 with Qubit 2
qc.cx(1, 2)
```

```python
# 5. Measure all 3 qubits and store them in the 3 classical bits.
qc.measure([0, 1, 2], [0, 1, 2])
```

```python
# Draw the circuit
print("Here is you 3-qubit Quantum Circuit:")
display(qc.draw("mpl"))
```

```python
# Run the simulation 2000 times
simulator = AerSimulator()
job = simulator.run(qc, shots=2000)
result = job.result()
```

```python
# Get and show the results
counts = result.get_counts(qc)
print(f"Results after 2000 runs: {counts}")
display(plot_histogram(counts))
```
